```python
def install_or_remove_packages():
    iOrR = ""
    while iOrR != "I" and iOrR != "R":
        print("Would you like to install or remove packages? (I/R)")
        iOrR = input().upper()
    if iOrR == "I":
        iOrR = "install"
    elif iOrR == "R":
        iOrR = "remove"
    return iOrR  # ← esto faltaba
```

```python
accion = install_or_remove_packages()
print(f"Se va a {accion} el paquete.")
```
