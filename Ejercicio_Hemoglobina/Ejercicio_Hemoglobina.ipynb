```python
# Importamos la libreria json para trabajar con archivos json
import json
```

```python
# ── Ejercicio 2: Cargar secuencia limpia ────────────────────

#- with open(...) abre un archivo de texto llamado hemoglobin_clean.txt.
# "r" significa modo lectura.
# f.read() lee todo el contenido del archivo.
# .strip() elimina espacios o saltos de línea al inicio y al final.
# El resultado se guarda en la variable sequence, que contiene la secuencia de aminoácidos (letras como A, C, D, etc.).


with open("hemoglobin_clean.txt", "r") as f:
    sequence = f.read().strip()
```

```python

# ── Ejercicio 4: Mostrar información de la secuencia ────────
# print("=" * 50) imprime 50 signos = para decorar.
# Se muestran datos básicos: nombre, organismo, longitud de la secuencia (len(sequence) cuenta cuántos aminoácidos hay).
# sequence[:30] muestra solo los primeros 30 aminoácidos.
# f"texto {variable}" es un f-string, que permite insertar variables dentro de un texto.

print("=" * 50)
print("  ANÁLISIS DE HEMOGLOBINA BETA HUMANA")
print("=" * 50)
print(f"Nombre de la proteína : Hemoglobin subunit beta")
print(f"Organismo             : Homo sapiens")
print(f"Longitud de secuencia : {len(sequence)} aminoácidos")
print(f"Secuencia             : {sequence[:30]}...")
print()
```

```python
# ── Ejercicio 5: Composición de aminoácidos ─────────────────
# Se define la lista de los 20 aminoácidos estándar.
# aa_count = {} crea un diccionario vacío.
# El for recorre cada aminoácido y cuenta cuántas veces aparece en la secuencia (sequence.count(aa)).
# Se guarda en el diccionario y se imprime.

aminoacids = [
    'A', 'C', 'D', 'E', 'F',
    'G', 'H', 'I', 'K', 'L',
    'M', 'N', 'P', 'Q', 'R',
    'S', 'T', 'V', 'W', 'Y'
]

print("Composición de aminoácidos:")
print("-" * 30)
aa_count = {}
for aa in aminoacids:
    count = sequence.count(aa)
    aa_count[aa] = count
    print(f"  {aa}: {count}")
print()

```

```python
# ── Ejercicio 6: Peso molecular (diccionario) ───────────────
molecular_weights = {
    'A':  89.09,  'C': 121.16,  'D': 133.10,  'E': 147.13,
    'F': 165.19,  'G':  75.03,  'H': 155.16,  'I': 131.17,
    'K': 146.19,  'L': 131.17,  'M': 149.21,  'N': 132.12,
    'P': 115.13,  'Q': 146.15,  'R': 174.20,  'S': 105.09,
    'T': 119.12,  'V': 117.15,  'W': 204.23,  'Y': 181.19
}

# Peso total = suma de pesos - agua liberada por cada enlace peptídico
raw_weight = sum(molecular_weights[aa] for aa in sequence)
molecular_weight = raw_weight - (len(sequence) - 1) * 18.02
print(f"Peso molecular estimado: {molecular_weight:.2f} Da")
print()

# molecular_weights es un diccionario con el peso de cada aminoácido en Daltons.
# sum(...) suma los pesos de todos los aminoácidos de la secuencia.
# Se resta (len(sequence)-1)*18.02 porque en cada enlace peptídico se libera una molécula de agua (18.02 Da).
# Se imprime el peso molecular aproximado.
```

```python
# ── Ejercicio 7: Función reutilizable ───────────────────────
# Se define una función que calcula el peso molecular de cualquier secuencia.
# seq: str indica que el parámetro es un string.
# -> float indica que devuelve un número decimal.
# Se usa igual que antes, pero ahora es reutilizable.

def calculate_molecular_weight(seq: str) -> float:
    """
    Calcula el peso molecular aproximado de una secuencia
    de aminoácidos (en Daltons).
    """
    weights = {
        'A':  89.09,  'C': 121.16,  'D': 133.10,  'E': 147.13,
        'F': 165.19,  'G':  75.03,  'H': 155.16,  'I': 131.17,
        'K': 146.19,  'L': 131.17,  'M': 149.21,  'N': 132.12,
        'P': 115.13,  'Q': 146.15,  'R': 174.20,  'S': 105.09,
        'T': 119.12,  'V': 117.15,  'W': 204.23,  'Y': 181.19
    }
    total = sum(weights[aa] for aa in seq if aa in weights)
    return total - (len(seq) - 1) * 18.02


# Luego se llama:

mw = calculate_molecular_weight(sequence)
print(f"Peso molecular (función): {mw:.2f} Da")
print()
```

```python
# ── Ejercicio 9: Aminoácidos hidrofóbicos ───────────────────
# Se define la lista de aminoácidos hidrofóbicos (repelen el agua).
# Se cuentan cuántos hay en la secuencia.
# Se calcula el porcentaje respecto al total.
# Se imprime el resultado.


hydrophobic = ['A', 'V', 'I', 'L', 'M', 'F', 'W', 'Y']
hydrophobic_count = sum(sequence.count(aa) for aa in hydrophobic)
hydrophobic_percent = (hydrophobic_count / len(sequence)) * 100

print(f"Aminoácidos hidrofóbicos: {hydrophobic_count} de {len(sequence)}")
print(f"Porcentaje hidrofóbico  : {hydrophobic_percent:.2f}%")
print()
```

```python

# ── Ejercicio 8: Guardar resultados en JSON ─────────────────
# Se crea un diccionario results con toda la información calculada.
# with open(..., "w") abre un archivo en modo escritura.
# json.dump(results, f, indent=4) guarda los resultados en formato JSON, con sangría de 4 espacios para que sea legible.
# Así se genera un archivo hemoglobin_results.json con todos los datos.


results = {
    "protein_name": "Hemoglobin subunit beta",
    "organism": "Homo sapiens",
    "sequence_length": len(sequence),
    "amino_acid_count": aa_count,
    "molecular_weight_Da": round(mw, 2),
    "hydrophobic_count": hydrophobic_count,
    "hydrophobic_percent": round(hydrophobic_percent, 2)
}

with open("hemoglobin_results.json", "w") as f:
    json.dump(results, f, indent=4)

print("✅ Resultados guardados en hemoglobin_results.json")
print("=" * 50)
```
